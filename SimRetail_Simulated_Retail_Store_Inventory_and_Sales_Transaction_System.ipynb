# **SimRetail: Simulated Retail Store Inventory and Sales Transaction System**

**DESCRIPTION**

 This Python script simulates a retail store's inventory and sales transactions. It utilizes various libraries like datetime, calendar, random, numpy, pandas, and uuid to generate realistic data.

**Key functionalities**:

1. Product catalog :  The script defines a dictionary named products that serves as the store's product catalog. Each product is represented as a list containing its price and current stock quantity.

2. Date and time simulation: The script leverages the datetime and calendar libraries to generate realistic date and time information for transactions.
Random customer selection: The script randomly selects customers from a predefined list using the random.choice function.

3. Product selection and quantity: The script simulates customers choosing random products and quantities using the random.choice and numpy.random.randint functions, ensuring that the selected quantity doesn't exceed the available stock.

4. Transaction details: For each transaction, the script generates a unique transaction ID using the uuid.uuid4 function and captures details like product name, price, quantity, date, and time.

5. Inventory update: After each transaction, the script updates the stock quantity of the purchased product in the products dictionary.

6. Dataframe creation: The script creates a Pandas dataframe named sales to store all transaction details for further analysis.

7. This script provides a basic framework for simulating retail store transactions and inventory management. It can be further extended to incorporate additional features like customer profiles, loyalty programs, discounts, and sales report.
